```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

url = "https://www.flipkart.com/apple-iphone-12-mini-black-64-gb/product-reviews/itm38b727191eb08?pid=MOBFWBYZXSEGBS6F&lid=LSTMOBFWBYZXSEGBS6FTONJ6W&marketplace=FLIPKART"
driver.get(url)

time.sleep(2)

# Creating empty list
page_urls=[]
complete_review=[]
stars=[]
start=0
end=350
time.sleep(2)

#scrapping all product Url

for page in range(start,end):#for loop for scrapping 350 pages
    url=driver.find_elements_by_xpath("//a[@class='ge-49M _2Kfbh8']")
    for i in url:
        page_urls.append(i.get_attribute('href'))
    nxt_button=driver.find_elements_by_xpath("//a[@class='_1LKTO3']")
    try:
        driver.get(nxt_button[1].get_attribute('href'))     #getting the link from the list for next page
    except:
        driver.get(nxt_button[0].get_attribute('href'))
            

for i in page_urls:
    driver.get(i)
    #for scrapping the number of stars
    for j in driver.find_elements_by_xpath("//div[@class='col _2wzgFH K0kLPL']/div[1]/div[1]"):
        stars.append(j.text)
    #for scrapping the complete review
    for l in driver.find_elements_by_xpath("//div[@class='t-ZTKy']/div/div"):
        complete_review.append(l.text)
               
        
#Combining all the lists into a single dataframe
df1=pd.DataFrame({'Number of ratings': stars,
               'Full Review': complete_review})
df1
```
